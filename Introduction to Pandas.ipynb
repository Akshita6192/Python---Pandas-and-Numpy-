```python
#Column from python list
fuel_economy = [7.5, 9.2, 5.0, 9.6, 8.7, 9.0, 8.0, 5.0, 9.0, 7.5]
car_sales["Fuel per 100km"] = fuel_economy 
car_sales
```

```python
#create column from another column
car_sales["total_fuel_used"] = car_sales["Odometer (KM)"]/100 * car_sales["Fuel per 100km"]
```

```python
#create column from a single value
car_sales["no of wheels"] = 4
car_sales
```
